```python
# なんもしてないデータ
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20, 6)

plt.title('g_rotate')
df3 = pd.read_json("https://ritoweb-public.s3.ap-northeast-1.amazonaws.com/normal_1.txt")
x = df3['x']
plt.plot(x, alpha=1)
n = 12
ave = np.convolve(x, np.ones(n)/float(n), 'same')
plt.plot(ave, alpha=1)
plt.legend(['before', 'after'],fontsize=12,loc=1)
```

```python
# ローパスフィルタ効いてる・・・？
# インターバルが0.018sだからだめなのかな。
plt.title('g_rotate_lopassfilter')
df3 = pd.read_json("https://ritoweb-public.s3.ap-northeast-1.amazonaws.com/lowpass_2.txt")
x = df3['y']
plt.plot(x, alpha=1)
n = 12
ave = np.convolve(x, np.ones(n)/float(n), 'same')
plt.plot(ave, alpha=1)
plt.legend(['x', 'x-move-ave'],fontsize=12,loc=1)
```
